```python
# Verificando as primeiras linhas do dataset
df2.head(5)
```


## Normalizando os dados


```python
#Normalizing the data with outliers
df_norm = preprocessing.normalize(df2)
df_norm = pd.DataFrame(df_norm)
```

```python
print(df_norm)
```

## Redução de dimensionalidade

```python
df_pca = PCA(n_components = 7)
df_pca2 = df_pca.fit_transform(df_norm)
```

```python
df_red = pd.DataFrame(data = df_pca2, columns = ['PC1','PC2','PC3','PC4','PC5','PC6','PC7'])
df_red 
```

```python
#Adicionando novamento o ID e TARGET
df_red ['ID'] = df['ID']
df_red ['TARGET'] = df['TARGET']
df_red  = df_red [['ID','PC1','PC2','PC3','PC4','PC5','PC6','PC7','TARGET']]
df_red 
```

## Contrução dos modelos

```python
#atribuindo a coluna TARGET a y e as colunas restantes a x.
x = df_red.drop(columns=['ID','TARGET'])
y = df_red['TARGET']

#conjunto de dados de treinamento e teste
x_treino, x_teste, y_treino, y_teste = train_test_split(x , y , test_size = 0.3 , random_state = 20)
x_treino.shape, x_teste.shape, y_treino.shape, y_teste.shape
```

```python
modelo_v1 = RandomForestClassifier(random_state = 42)
modelo_v1.fit(x_treino, y_treino.ravel())
```

```python
#Treinando
# Verificando os dados de treino
rf_predict_train = modelo_v1.predict(x_treino)
print("Exatidão (Accuracy): {0:.4f}".format(metrics.accuracy_score(y_treino, rf_predict_train)))
```

```python
# Verificando nos dados de teste
rf_predict_test = modelo_v1.predict(x_teste)
print("Exatidão (Accuracy): {0:.4f}".format(metrics.accuracy_score(y_teste, rf_predict_test)))
print()
```

# Balanceamento dos dados

## Resampling = Reamostragem


```python
def under_sampler(N,x = df.TARGET.value_counts()[1]):
    class_0 = df_red[df_red['TARGET']==0].sample( int(N*x) ,random_state=1)
    class_1 = df_red[df_red['TARGET']==1]
    return pd.concat([class_0,class_1],axis=0)

df_0 = under_sampler(4)
df_0.shape
```

```python
df_0['TARGET'].value_counts()
```

```python
df_0
```

```python
#atribuindo a coluna TARGET a y e as colunas restantes a x.
x = df_0.drop(columns=['ID','TARGET'])
y = df_0['TARGET']

#conjunto de dados de treinamento e teste
x_treino, x_teste, y_treino, y_teste = train_test_split(x , y , test_size = 0.3 , random_state = 20)
x_treino.shape, x_teste.shape, y_treino.shape, y_teste.shape
```

```python
modelo_v2 = RandomForestClassifier(random_state = 42)
modelo_v2.fit(x_treino, y_treino.ravel())
```

```python
# Verificando os dados de treino
rf_predict_train = modelo_v2.predict(x_treino)
print("Exatidão (Accuracy): {0:.4f}".format(metrics.accuracy_score(y_treino, rf_predict_train)))
```

```python
# Verificando nos dados de teste
rf_predict_test = modelo_v2.predict(x_teste)
print("Exatidão (Accuracy): {0:.4f}".format(metrics.accuracy_score(y_teste, rf_predict_test)))
print()
```

# Resampling : Over-Samplin

Resampling : Over-Sampling  Geração de casos adicionais (cópias, observações artificiais) da classe minoritária para
aumentar seu efeito sobre o classificador. Todos os casos da classe majoritária são mantidos.

#Repeticao
Vamos multiplicar a quantidade de classe 1 pontos por dois. O problema com essa abordagem é que ela corre o risco de se 
adaptar demais, uma vez que o classificador vê os mesmos dados repetidamente (para ser mais preciso, ele vê duas vezes).



```python
df_1 = df_red.copy()

df_1 = pd.concat([df_1,df_1[df_1.TARGET == 1]],axis=0)
df_1.TARGET.value_counts()
```

```python
#atribuindo a coluna TARGET a y e as colunas restantes a x.
x = df_1.drop(columns=['ID','TARGET'])
y = df_1['TARGET']

#conjunto de dados de treinamento e teste
x_treino, x_teste, y_treino, y_teste = train_test_split(x , y , test_size = 0.3 , random_state = 20)
x_treino.shape, x_teste.shape, y_treino.shape, y_teste.shape
```

```python
# Verificando os dados de treino
rf_predict_train = modelo_v1.predict(x_treino)
print("Exatidão (Accuracy): {0:.4f}".format(metrics.accuracy_score(y_treino, rf_predict_train)))
```
